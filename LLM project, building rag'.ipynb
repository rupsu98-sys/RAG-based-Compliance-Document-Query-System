```python
def chunk_text(text, chunk_size=512, overlap=128):
    """Chunks text with overlap to maintain context."""
    words = text.split()
    chunks = []
    for i in range(0, len(words), chunk_size - overlap):
        chunk = " ".join(words[i : i + chunk_size])
        chunks.append(chunk)
    return chunks
```

```python
def create_documents(chunks, source):
    """Converts text chunks to LlamaIndex document objects with metadata."""
    return [Document(text=chunk, metadata={"source": source}) for chunk in chunks]
```

```python
import faiss
from transformers import AutoTokenizer, AutoModel
import numpy as np

class FAISSIndex:
    def __init__(self, dimension=384):  # Changed dimension to 384
        self.index = faiss.IndexFlatL2(dimension)
        self.docs = []
        # Import AutoTokenizer and AutoModel here
        self.tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")
        self.model = AutoModel.from_pretrained("sentence-transformers/all-MiniLM-L6-v2")

    def encode_text(self, text):
        inputs = self.tokenizer(text, return_tensors='pt', padding=True, truncation=True)
        outputs = self.model(**inputs)
        return outputs.pooler_output.detach().numpy()[0]

    def add_documents(self, documents):
        embeddings = np.array([self.encode_text(doc.text) for doc in documents]).astype('float32')

        # Check the actual dimension of the embeddings
        embedding_dimension = embeddings.shape[1]

        # If the dimension doesn't match the index dimension, raise a more informative error
        if embedding_dimension != self.index.d:
            raise ValueError(f"Embedding dimension ({embedding_dimension}) does not match index dimension ({self.index.d}). "
                             f"Please initialize FAISSIndex with the correct dimension.")

        self.index.add(embeddings)
        self.docs.extend(documents)

    def search(self, query, top_k=3):
        query_embedding = self.encode_text(query).reshape(1, -1)
        _, indices = self.index.search(query_embedding, top_k)
        return [self.docs[i] for i in indices[0]]
```

```python
def build_index(documents):
    """Builds an index and stores embeddings in FAISS."""
    index = FAISSIndex(dimension=384)  # Changed dimension to 384 to match the embedding dimension
    index.add_documents(documents)
    return index
```

```python
if __name__ == "__main__":
    file_path = input("Enter file path: ").strip()
    all_documents = []

    text = read_text_from_file(file_path)
    if text:
        chunks = chunk_text(text)
        documents = create_documents(chunks, source=file_path)
        all_documents.extend(documents)

    index = build_index(all_documents)
    query_engine = get_query_engine(index)

    while True:
        query = input("Enter your query (or type 'exit' to quit): ")
        if query.lower() == "exit":
            break
        response = query_engine(query)
        print("Response:", response)
```
